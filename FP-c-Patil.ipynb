## Description of the CSV dataset you download (one paragraph):
- This dataset shows the Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs)

## Display all fields information (column name with short description per line):
- VIN(1-10):Vehicle Identification Number
- County: Count of the car.
- City: City of the car.
- State: State of the car.
- Postal Code: Postal code of the car, basically the addressof the car.
- Model Year: this column suggest the model year, means model from which year.
- Make: this column suggest that the manufacturer of the car(BMW, NISSAN TESLA, HYUNDAiI etc.)
- Model: Model of the car.
- Electric Vehicle Type: whether the car is Plug-in Hybrid Electric Vehicle (PHEV), Battery Electric Vehicle (BEV).
- Clean Alternative Fuel Vahicle Eligibility:
- Electric Range: range of electricity.
- Base MSRP: Manufacturer Suggested Retail Price, but it showing 0 in data.
- Legislative District: law of the dictrict.
- DOL Vehicle ID: Department of Licensing, ID.
- Vehicle Location: location of the vehicle.
- Electric Utility: Utilities for electricity.
- 2020 Census Tract: Census tracking number.
